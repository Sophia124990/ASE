```python
#Generamos un DataFrame con los datos de los cuatro archivos
import pandas as pd

#Leemos los archivos CSV
pobF_df = pd.read_excel('/content/PobTotalPaisFem.xlsx')
pobM_df = pd.read_excel('/content/PobTotalPaisMas.xlsx')
```

```python
# Unir los DataFrames en la columna 'COUNTRY'
df_unido = pd.merge(pobF_df, pobM_df, on='COUNTRY', how='inner')
df_unido.head()
```

```python
import matplotlib.pyplot as plt
# Convertir las columnas a tipo numérico, manejando posibles errores
# The '2014F' column is likely already numeric or of a type that doesn't support string operations.
# Remove the .str.replace to directly convert the column to numeric type.
df_unido['2014F'] = pd.to_numeric(df_unido['2014F'], errors='coerce')
df_unido['2014M'] = pd.to_numeric(df_unido['2014M'], errors='coerce')
df_unido['2015F'] = pd.to_numeric(df_unido['2015F'], errors='coerce')
df_unido['2015M'] = pd.to_numeric(df_unido['2015M'], errors='coerce')
df_unido['2016F'] = pd.to_numeric(df_unido['2016F'], errors='coerce')
df_unido['2016M'] = pd.to_numeric(df_unido['2016M'], errors='coerce')

df_unido['2017F'] = pd.to_numeric(df_unido['2017F'], errors='coerce')
df_unido['2017M'] = pd.to_numeric(df_unido['2017M'], errors='coerce')
df_unido['2018F'] = pd.to_numeric(df_unido['2018F'], errors='coerce')
df_unido['2018M'] = pd.to_numeric(df_unido['2018M'], errors='coerce')
df_unido['2019F'] = pd.to_numeric(df_unido['2019F'], errors='coerce')
df_unido['2019M'] = pd.to_numeric(df_unido['2019M'], errors='coerce')

df_unido['2020F'] = pd.to_numeric(df_unido['2020F'], errors='coerce')
df_unido['2020M'] = pd.to_numeric(df_unido['2020M'], errors='coerce')
df_unido['2021F'] = pd.to_numeric(df_unido['2021F'], errors='coerce')
df_unido['2021M'] = pd.to_numeric(df_unido['2021M'], errors='coerce')
df_unido['2022F'] = pd.to_numeric(df_unido['2022F'], errors='coerce')
df_unido['2022M'] = pd.to_numeric(df_unido['2022M'], errors='coerce')

df_unido['2023F'] = pd.to_numeric(df_unido['2023F'], errors='coerce')
df_unido['2023M'] = pd.to_numeric(df_unido['2023M'], errors='coerce')
# Especificar las columnas que quieres incluir en el boxplot
columnas_a_graficar = ['2014F','2014M', '2015F', '2015M', '2016F', '2016M', '2017F','2017M', '2018F', '2018M', '2019F', '2019M', '2020F', '2020M', '2021F', '2021M', '2022F', '2022M', '2023F', '2023M']

# Crear el boxplot
plt.figure(figsize=(20, 7))
df_unido.boxplot(column=columnas_a_graficar)
plt.title('Boxplot de población F y M por año de 2014 a 2023')
plt.xlabel('Año')
plt.ylabel('Valores')
plt.show()
```
